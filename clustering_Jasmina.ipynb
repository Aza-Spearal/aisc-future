```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
vecs = np.load('activations2.npy') #Loads activations from layer 12
```

```python
np.shape(vecs)
```

```python
def cossim(x,y):
    return np.dot(x,y)/(np.linalg.norm(x)*np.linalg.norm(y))
```

```python
def plotfunc(v):
    vp=np.zeros([int(len(v[:,0])),int(len(v[:,0]))])
    for i in range (len(vp[:,0])):
        for j in range (len(vp[0,:])):
            vp[i,j]=cossim(v[i,:],v[j,:])
    ax=sns.heatmap(vp, square=True, center=0.0)
    plt.show()
```

```python
plotfunc(vecs)
```
